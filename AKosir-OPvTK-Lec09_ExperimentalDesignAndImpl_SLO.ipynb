## ■ Kaj je načrt poskusa

Elementi načrta poskusa so naslednji:
##### 1. Odvisna spremenljivka (odziv, angl. response) $y=c$: 
Kriterijska funkcija, indikator uspešnosti ipd, ki podaja glavni rezultat. V našem primeru uspešnost pogovora. 

##### 2. Kontroliran(i) faktor(ji) $F_e$: 
Neodvisne spremenljivke, ki jih v poskusu **kontroliramo**. V našem primeru je to komunikacijska naprava, torej ima kontroliran faktor vrednosti (nivoje) {PC, Alexa Echo}.   

##### 3. Moteči (angl. nuisance) faktorji: 
To so nezaželjeni vplivi na eksperimentalni rezultat, ki so neizogibni. Želimo jih nevtralizirat na več načinov;

##### 4. Vhodni šum $\varepsilon$:
To je šum, s katerim predstavimo nepoznane moteče faktorje in resnični šum, ki izhaja naključnosti človeškega obnašanja, šuma senzorjev ipd.

 - dvorazsežna (angl. two way): načrt ima dva kontrolirana faktorja

## ■ Načrt ANOVA 

$\def\ovr#1{{\overline{#1}}}$
$\def\o#1{{\overline{#1}}}$
$\def\SS{{SS}}$
$\def\Var{{Var}}$


#### Kontrolirani in moteči faktorji

Vrste:
- fiksen učinek (angl. fixed effect): eksperimentalni faktor kontroliramo tako, da mu določimo vrednost pred izvedbo poskusa 
- slučajen (angl. random effect): vrednost eksperimentalnega faktorja je naključna, to je vzorčimo iz poppulacije in le beležimo vrednosti faktorja
- model z mešainmi učinki (angl. mixed effect): imamo faktorje s fiksnimi in slučajnimi učinki


Zaloga vrednosti: 
- diskretne vrednosti
- dve ali več vrednosti

#### Pogoji

Spremenljivke so porazdeljene normalno. 



#### Ničelna hipoteza in test

Pokrijemo le varianto za **en faktor** s **fiksnim učinkom**. Faktor ima $a$ različnih vrednosti, vrednosti faktorja so indeksirani kot $\{1, 2, \ldots, a\}$ in po teh vrednostih se grupirajo rezultati v $\{G_1, \ldots, G_a\}$

Ničelna hipoteza:
$$ H_{0} = [\ovr{y}_{G_1} = \cdots = \ovr{y}_{G_a}],  $$

Test: F-test.



**Pri tem je:**

###### Oznake
- $n$ je število izvedb poskusa pri vsaki vrednosti kontroliranega faktorja $F_e$. 
- $a$ je število vrednosti kontroliranega faktorja $F_e \in \{1, 2, \ldots, a\}$
- $y_{ij}$ je $j$-ti rezultat poskusa, $i=1,\ldots, n$
- $y_{i.} = \sum_{j=1}^n y_{ij}$ vsote po poskusih
- $\o{y}_{i.} = y_{i.}/n$ povprečje po poskusih
- $y_{.j} = \sum_{i=1}^a y_{ij}$ vsote po vrednostih faktorja
- $\o{y}_{.j} = y_{.j}/n$ povprečje po vrednostih faktorja
- $y_{..} = \sum_{i,j} y_{ij}$ celotna vsota
- $\o{y}_{..} = y_{..}/n$ celotno povprečje

##### Vsote kvadratov
- totalna vsota kvadratov,  vsota kvadratov faktorja, vsota vsot:
$$ \SS_{tot} = \sum_{i=1}^a \sum_{j=1}^n (y_{ij} - \o{y_{..}})^2, \qquad
   \SS_{trt} = n \sum_{i=1}^a (\o{y}_{i.} - \o{y_{..}})^2, \qquad
   \SS_{tot} = \SS_{trt} + \SS_{err}  $$


##### Statistika F

Statistika
$$ F_0 = \frac{S_{trt}/(a-1)}{SS_{err}/(n(a-1))} $$

je porazdeljena po porazdelitvi $F(a-1, n(a-1))$, od koder izračunamo $p$ vrednost kot 

$$ p = 1 - F(a-1, n(a-1)).{cdf}(F_0),  $$

kjer je $F(a-1, n(a-1)).{cdf}$ komulativna porazelitvena funkcija $F$ porazdelitve $F(a-1, n(a-1)$.

<div style="width:100%;text-align:right;font-weight:bold;font-size:1.2em;"> 9 </div>

```python
# Define functions
def get_ANOVA_SS(data_df):
    a,n = data_df.shape
    N = a*n
    
    y_pp = 1.0*data_df.sum().sum()
    y_ip = 1.0*data_df.sum(axis=1)
    
    # SS total
    SS_T = (data_df**2).sum().sum() - (y_pp**2)/N

    # SS treatment
    SS_trt = (1.0/n)*(y_ip**2).sum() - (y_pp**2)/N
    
    # SS error
    SS_err = SS_T - SS_trt
   
    # Report
    return SS_trt, SS_err, SS_T
```

```python
## ANOVA

import numpy as np
import pandas as pd 
from scipy.stats import f

# Load data 
post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/ANOVAtestData31.csv'
data_df = pd.read_csv(post_fn, sep=';', encoding='utf8')


# Summs of squares
SS_trt, SS_err, SS_T = get_ANOVA_SS(data_df)

print (SS_trt, SS_err, SS_T)

# F-stat
a,n = data_df.shape
N = a*n
F_0 = (SS_trt/(a-1)) / (SS_err/(N-a))

# P-value
pVal = 1-f.cdf(F_0, a-1, N-a)

print ('p value = ', pVal)
```

## ■ Določanje Velikost vzorca

Določimo s pomočjo a-priori analize moči, pri kateri iz povezave med močjo (občutljivostjo testa) $P[\neg\hat{H_0}\big|_{\neg H_0}]$ in velikostjo vzorca $n$ pri izbrani občutljivosti določimo velikost vzorca.

#### Velikost učinka (angl. effect size)

To je mera za velikost odklona od ničelne hipoteze $H_0$ in je definirana za vsak test posebej. Za ANOVA pri dveh skupinag $G_1, G_2$ je dana kot razlika med povprečji grup, deljena s skupnim standarnim odklonom $\sigma_p$. 
$$ ef = \frac{\overline{y}_{G_1} - \overline{y}_{G_2}}{\sigma_p} $$

Pri določanju velikosti vzorca vrednosti velikosti učinka ne poznamo in ga ocenimo ali iz sorodnih ali preteklih študij ali drugih virov. Pri tem ločimo
- majhen: $ef < 0.2$
- srednji: $ef \in [0.2, 0.4]$
- velik: $ef > 0.4$ 
<p>
Različni viri navajajo različne vrednosti.

#### Privzeta moč testa:

Moč testa $1-\beta = P[\neg\hat{H_0}\big|_{\neg H_0}]$, ki je verjetnost, da ničelno hipotezo zavrnemo kot ta ne drži, privzamemo vnaprej. 
Smiselne vrednosti so med $0.8$ in $0.95$. 

Višje privzete moči testa zahtevajo večje vzorce (več udeležencev testiranj).


#### Uporaba oroda GPower 

Orodje GPower (https://www.psychologie.hhu.de/arbeitsgruppen/allgemeine-psychologie-und-arbeitspsychologie/gpower) nam omogoča izvedbo določanja velikosti vzorca.  

Za ANOVA postavitev za A/B testiranja so nastavitve na spodnji sliki:
- Test family: F-test
- Statistical test: ANOVA, Fixed effect, omnibus, one way (tu lahko izberemo druge ustrezne možnosti, da je tipična za A/B testiranja)
- Type of power analysis: apriori 
- Parametri: 
    - Effect size: preizkusimo različne vrednosti med 0.2 in 0.5
    - $\alpha$ error prob: stopnja tveganja je $0.05$
    - Power: izbiramo med $0.8$ do $0.95$
    - Number of groups: število skupin, pri A/B je to 2, pri A/B/C je 3 itd.

## ■ Korak 4: Določitev eksperimentalnega scenarija


#### Def.: Eksperimentalni scenarij je scenarij, ki ga opravijo testne osebe z namenom pridobitve relevantnih meritev in rezultatov


#### Zamisljiv cilj ekperimenta

- cilj eksperimenta je **cilj, ki ga imajo testne osebe v mislih med izvedbo eksperimenta** in ne raziskovalni cilj eksperimenta!
- daje **referenčni okvir** rezulatov eksperimenta - meritev in odgovorov na vprašanja
- neposredno je povezan z
  - eksperimentalno storitvijo 
  - eksperimentalno vsebino


#### Vodila za eksperimentalni scenarij

- eksperimentalni scenarij je zamisljiv za testne osebe:
  - starejši 60+ in "escape room" ne gresta skupaj
  - ...


- eksperimentalni scenarij mora biti dovolj preprost, da
  - ga je mogoče predstaviti z navodili
  - testne osebe ne doživijo **nenačrtovanih presenečenje** - vsako presenečenje uniči refeenčni okvir testne osebe in poveča nezanesljivost odzivov (odgovorov in psihofizioloških meritev)


- pomembna so jasna navodila in vodila, ki testne osebe ne vodijo v **presenečenja** med izvedbo  


- vsaka **možnost razvoja dogodkov** je možnost za nov podatek o uporabniku

- primeren kognitivni napor testnih oseb:
  - premajhen: nezainteresirani

#### Eksperimentalni načt: testiranje hipotez

Eksperimentalni načrt da podatke za **testiranje hipotez**
- iz **raziskovalnega vprašanja** oblikujemo **delovno hipotezo**
  - naš primer pogovora z Alexo: 
    - raziskovalno vprašanje: ali vmesnik (Echo ali simulator) vpliva da percepcijo kvalitete dvogovora?
    - delovna hipoteza: vmesnik vpliva na percepcijo kvalitete dvogovora
- na podlagi delovne hipoteze oblikujemo **ničelno hipotezo**
  - naš primer pogovora z Alexo:
    - ničelna hipoteza(e)
    $$ H_0 = [\overline{y}_S = \overline{y}_E], $$
    kjer je $\overline{y}_S$ povprečje rezultatov skupine s simulatorjem in $\overline{y}_E$ povprečje rezultatov skupine z napravo Echo.

```python
# Define functions
def get_ANOVA_SS(data_df):
    
    factor_l = data_df['G'].unique()

    y = data_df['y']
    y_pp = data_df['y'].mean()
    y_ip = [data_df[data_df['G']==ii]['y'].mean() for ii in factor_l]

    SS_T = ((y - y_pp)**2).sum()
    SS_trt = ((y_ip - y_pp)**2).sum()
    SS_err = SS_T - SS_trt
    
    # Report
    return SS_trt, SS_err, SS_T
```

```python
# one-way ANOVA fixed factor
import pandas as pd 
from scipy.stats import f



latent_fs = ['UGO', 'CSB', 'SC', 'V']
groups = dict.fromkeys([0,1,2,3,4,5,6,7], 1) # Echo
groups.update(dict.fromkeys([8,9,10,11,12,13,14,15], 2)) # PC

# For all vars
for fs_n in latent_fs:

    curr_df = pd.DataFrame(columns=['G', 'y'])
    curr_df['G'] = groups
    curr_df['y'] = speech_quality_est_df[fs_n]
    

    
    # Compute SS
    #curr_df = pd.concat([sq_est_PC, sq_est_Echo]) # pd.DataFrame([sq_est_PC.as_matrix(), sq_est_Echo.as_matrix()]) #, columns = [1,2,3,4,5,6,7,8])
    SS_trt, SS_err, SS_T = get_ANOVA_SS(curr_df)
    

    # F-stat
    a = len(curr_df['G'].unique())
    N = curr_df.shape[0]
    F_0 = (SS_trt/(a-1)) / (SS_err/(N-a))
    
    # P-value
    p_val = 1-f.cdf(F_0, a-1, N-a)
    
    # Report
    print ('Latent var:', fs_n, 'p-val=', p_val)
    
```
