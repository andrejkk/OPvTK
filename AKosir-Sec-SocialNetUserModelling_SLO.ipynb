# ■ Merjenje centralnosti - teorija in primer

1. Centralnost stopenj (angl. Degree centralitiy)

Centralnost stopnje meri, koliko neposrednih povezav ima vozlišče.
$$ C_D(v) = \frac{deg(v)}{N - 1} $$


2. Medsebojna centralnost (angl. Betweenness Centrality)

Medsebojna centralnost meri, kako pogosto vozlišče deluje kot most med drugimi vozlišči.
$$ C_B(v) = \sum_{s \neq v \neq t} \frac{\sigma_{st}(v)}{\sigma_{st}} $$
kjer
$\sigma_{st}(v)$ = število najkrajših poti med vozlišči $s$ in $t$ in $\sigma_{st}$ = število najkrajših poti skozi 


3. Centralnost bližine (angl. Closeness Centrality)

Centralnost bližine meri, kako blizu je vozlišče vsem drugim vozliščem v grafu.
$$ C_C(v) = \frac{N-1}{\sum_{u \neq v} d(v, u)} $$


4. Centralnost lastnih vektorjev (angl. Centrality)

Eigenvektorska centralnost meri vpliv vozlišča z upoštevanjem pomembnosti njegovih sosedov.

$$ C_E(v) = \frac{1}{\lambda} \sum_{u \in N(v)} C_E(u) $$


5. Razgradnja na  k-jedra (angl. K-Core Decompositio)

Vozlišče pripada k-jedru, če ima vsaj k povezav.

$$ K(v) = \max \{ k \mid v \text{ pripada k-jedru} \} $$

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors

# Step 1: Generate a random graph (Erdős-Rényi model)
num_nodes = 15  # Number of nodes
prob_edge = 0.3  # Probability of edge creation
G = nx.erdos_renyi_graph(num_nodes, prob_edge, seed=42)

# Step 2: Compute centrality measures
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
k_core = nx.core_number(G)

# Step 3: Print centrality measures
print("Degree Centrality:", degree_centrality)
print("Betweenness Centrality:", betweenness_centrality)
print("Closeness Centrality:", closeness_centrality)
print("K-Core Decomposition:", k_core)

# Step 4: Visualize the graph with centrality measures
fig, ax = plt.subplots(figsize=(8, 6))  # Create a figure and axis
pos = nx.spring_layout(G, seed=42)  # Layout for visualization

# Normalize node colors for color mapping
node_color_values = list(degree_centrality.values())
norm = colors.Normalize(vmin=min(node_color_values), vmax=max(node_color_values))
cmap = cm.viridis

# Draw nodes with color based on degree centrality
nodes = nx.draw_networkx_nodes(G, pos, ax=ax, node_size=700, 
                               node_color=node_color_values, cmap=cmap, edgecolors="black")
nx.draw_networkx_edges(G, pos, ax=ax, edge_color="gray")
nx.draw_networkx_labels(G, pos, ax=ax, font_size=10, font_color="black")

# Add colorbar properly
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # Dummy array for colorbar
cbar = plt.colorbar(sm, ax=ax, label="Degree Centrality")

plt.title("Random Graph with Centrality Measures (Degree Centrality)")
plt.show()
```

# ■ Določanje skupnosti

Cilj določanja skupnosti je poiskati podgrafe (skupnosti) v grafu, kjer imajo vozlišča (uporabniki) veliko povezav znotraj svoje skupine, vendar manj povezav z drugimi skupinami.

### 1. Metoda Louvain (optimizacija modularnosti)

Začne se z vsakim vozliščem kot lastno skupnostjo. Iterativno združuje skupnosti, da bi povečal modularnost, ki je merilo, kako dobro so skupnosti opredeljene.
Ustavi se, ko modularnosti ni mogoče več povečati.

Formula za modularnost, ki jo algortiem maksimizira, je
$$ Q(G) = \frac{1}{2m} \sum_{u,v}\left(A_{uv} - \frac{\deg(u)\deg(v)}{2 m}\right)\delta(c_u, c_v), $$
kjer 
 - $m=|E|$, 
 - $A_{uv}$ je matrika pripadnosti grafa, 
 - $\deg(u)$ je stopnja vozlišča $u$ in 
 - $\delta(c_u, c_v)$ je indikatorska funkcija ($1$, če sta $u,v$ v isti skupnosti, in $0$ v nasprotnem primeru).


### 2. Metoda Girvan-Newmanova

Ta metoda izračuna središčnost vmesnosti vseh robov (to je, kako pogosto se rob pojavi na najkrajših poteh) in odstrani rob z največjo vmesnostjo. Postopek se ponavlja, dokler graf ne razpade na ločene podgrafe.


### 3. Razširjanje oznak

Vsakemu vozlišču dodeli naključne oznake. Vsako vozlišče posodobi svojo oznako, da se ujema z najpogostejšo oznako njegovih sosedov.
Konvergira, ko ima večina vozlišč stabilne oznake.

# ■ Razširjanje vpliva

Širjenje vpliva modelira, kako se ideje, trendi ali vedenje širijo med uporabniki v družbenem omrežju.

### 1. Model neodvisne kaskade (IC)

Ideja je, da ima vsak uporabnik, na katerega je bil dosežen vpliv, enkratno priložnost, da aktivira svoje sosede. Na začetku se aktivira začetni nabor uporabnikov.
Aktivirani uporabniki poskušajo vplivati na svoje sosede z verjetnostjo $p$. Postopek se nadaljuje, dokler ni več aktivacij.

Verjetnost aktivacije je 
$$ P[v \text{ is activaded by } u] = p_{uv} $$
je določena z različnimi pristopi.

#### 2. Model linearnega praga (LT)

Po tem pristopu uporabnik sprejme zamisel, če vsota vplivov sosedov preseže prag. Prag $T_u$ je dodeljen vsakemu uporabniku in uporabnik je pod vplivom, če vsota prejetih vplivov preseže njegov prag. Postopek se nadaljuje, dokler spremembe niso dovolj majhne. 

$$ \sum_{u\in N(v)} w_{uv} \geq T_u. $$

### 3. Model dovzetnosti, okužbe in ponovnega odkritja (SIR)

Zamisel izhaja iz epidemičnega širjenja, uporabniki so lahko:
- dovzetni (S) - še niso okuženi.
- okužen (I) - trenutno širi vpliv.
- ozdravljen (R) - ne širi več vpliva.

Algoritem izvede naslednje korake:
1. Na dovzetne uporabnike vplivajo njihovi sosedje z verjetnostjo $p$
2. Okuženi uporabniki širijo vpliv, vendar si sčasoma opomorejo (prenehajo širiti vpliv).

```python
import networkx as nx
import random

def independent_cascade(G, seed_nodes, p=0.2, steps=5):
    """
    Simulates influence propagation using the Independent Cascade Model.
    
    Parameters:
        G (nx.Graph): The social network graph.
        seed_nodes (list): Initial set of influenced nodes.
        p (float): Probability of influence propagation.
        steps (int): Number of iterations for propagation.
    
    Returns:
        activated_nodes (set): Nodes influenced at the end of propagation.
    """
    activated_nodes = set(seed_nodes)
    newly_activated = set(seed_nodes)
    
    for _ in range(steps):
        next_activated = set()
        for node in newly_activated:
            for neighbor in G.neighbors(node):
                if neighbor not in activated_nodes and random.random() < p:
                    next_activated.add(neighbor)
        
        if not next_activated:
            break
        
        activated_nodes.update(next_activated)
        newly_activated = next_activated
    
    return activated_nodes

# Create a sample social network graph
G = nx.erdos_renyi_graph(30, 0.15, seed=42)  # Random graph with 20 nodes
seed_nodes = [0]  # Start with node 0 as the influenced user

# Run influence propagation
final_influenced_nodes = independent_cascade(G, seed_nodes, p=0.2, steps=5)

print("Final influenced nodes:", final_influenced_nodes)

# Visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
color_map = ['red' if node in final_influenced_nodes else 'blue' for node in G.nodes()]
nx.draw(G, with_labels=True, node_color=color_map, node_size=500, edge_color='gray')
plt.show()
```
