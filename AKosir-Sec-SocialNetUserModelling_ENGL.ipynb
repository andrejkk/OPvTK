# ■ Merjenje centralnosti - teorija in primer

1. Degree Centrality

Degree centrality measures how many direct connections a node has.
$$ C_D(v) = \frac{deg(v)}{N - 1} $$


2. Betweenness Centrality

Betweenness centrality measures how often a node acts as a bridge between other nodes.
$$ C_B(v) = \sum_{s \neq v \neq t} \frac{\sigma_{st}(v)}{\sigma_{st}} $$
where
$\sigma_{st}(v)$ = Number of shortest paths between nodes $s$ amd $t$ and $\sigma_{st}$ = Number of shortest paths passing through 


3. Closeness Centrality

Closeness centrality measures how close a node is to all other nodes in the graph.
$$ C_C(v) = \frac{N-1}{\sum_{u \neq v} d(v, u)} $$


4. Eigenvector Centrality

Eigenvector centrality measures the influence of a node by considering its neighbors’ importance.

$$ C_E(v) = \frac{1}{\lambda} \sum_{u \in N(v)} C_E(u) $$


5. K-Core Decomposition

A node belongs to a k-core if it has at least k connections.

$$ K(v) = \max \{ k \mid v \text{ belongs to the k-core} \} $$

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors

# Step 1: Generate a random graph (Erdős-Rényi model)
num_nodes = 15  # Number of nodes
prob_edge = 0.3  # Probability of edge creation
G = nx.erdos_renyi_graph(num_nodes, prob_edge, seed=42)

# Step 2: Compute centrality measures
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
k_core = nx.core_number(G)

# Step 3: Print centrality measures
print("Degree Centrality:", degree_centrality)
print("Betweenness Centrality:", betweenness_centrality)
print("Closeness Centrality:", closeness_centrality)
print("K-Core Decomposition:", k_core)

# Step 4: Visualize the graph with centrality measures
fig, ax = plt.subplots(figsize=(8, 6))  # Create a figure and axis
pos = nx.spring_layout(G, seed=42)  # Layout for visualization

# Normalize node colors for color mapping
node_color_values = list(degree_centrality.values())
norm = colors.Normalize(vmin=min(node_color_values), vmax=max(node_color_values))
cmap = cm.viridis

# Draw nodes with color based on degree centrality
nodes = nx.draw_networkx_nodes(G, pos, ax=ax, node_size=700, 
                               node_color=node_color_values, cmap=cmap, edgecolors="black")
nx.draw_networkx_edges(G, pos, ax=ax, edge_color="gray")
nx.draw_networkx_labels(G, pos, ax=ax, font_size=10, font_color="black")

# Add colorbar properly
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # Dummy array for colorbar
cbar = plt.colorbar(sm, ax=ax, label="Degree Centrality")

plt.title("Random Graph with Centrality Measures (Degree Centrality)")
plt.show()
```

# ■ Razširjanje vpliva

Influence propagation models how ideas, trends, or behaviors spread among users in a social network.

### 1. (A) Independent Cascade (IC) Model

The idea is that each influenced user has a one-time chance to activate their neighbors. Initially, a seed set of users is activated.
Activated users try to influence their neighbors with probability $p$. The process continues until no more activations occur.

Probability of activation is 
$$ P[v \text{is activaded by } u] = p_{uv} $$
is determined by different approaches.

### 2. (B) Linear Threshold (LT) Model

The approach is a user adopts an idea if the sum of influences from neighbors exceeds a threshold. A threshold $T_u$ is assigned to each user and a user is influenced if the sum of incoming influences crosses her threshold. The process continues until changes are small enough. 

$$ \sum_{u\in N(v)} w_{uv} \geq T_u. $$

### 3. (C) Susceptible-Infected-Recovered (SIR) Model

The idea comes from epidemic spread, users can be:
- Susceptible (S) – Not yet influenced.
- Infected (I) – Currently spreading the influence.
-  Recovered (R) – No longer spreading influence.

The algorithm executes the following steps:
1. Susceptible users get influenced by their neighbors with probability $p$
2. Infected users spread the influence but eventually recover (stop spreading).
