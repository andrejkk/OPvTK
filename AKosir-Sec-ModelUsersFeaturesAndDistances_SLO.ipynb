## ■ Zakaj razdalje in podobnosti

Eden od glavnih mehanizmov prilagajanja je prilagajanje **na najbolje rangirno** prilagoditev ali vsebino. 

#### Kaj je 
- **Razdalja**: matematična funkcija na vektorju lastnosti, ki meri razdaljo
- **Podobnost**: matematična funkcija na vektorju lastnosti, ki meri podobnost

Pri določanju vrstnega reda v postopku rangirajo uporabjamo **razdalje** in **podobnosti** med 
uporabniki in med vsebinami. Zato bomo razdalje, podobnosti in rangoranja podrobneje obravnavali. 

#### Od kod lastnosti uporabnikov in vsebin:
- Demografski podatki uporabnikov: starost, 
- Zgodovina uporabe - model: katere filme je že videl, žanri teh filmov, ...
- Konteksti uporabe
- Osebnost, razpoloženje in čustva uporabnika
- **latentne** spremenljivke: lastnosti iz matrične faktorizacije, lastnosti iz redukcije dimenzije

## ■ Kaj je razdalja (metrika)


Razdalja (metrika) na množici $X$ je funkcija
$$ d: X\times X \to ℝ_+$$ 
ki zadošča zahtevam
- nenegativnost: $d(x_1, x_2)\geq 0$
- identičnost: $d(x_1, x_2) = 0 \Rightarrow x_1 = x_2$
- simetričnost: $d(x_1,x_2)=d(x_2,x_1)$
- trikotniška neenakost: $d(x_1, x_3) \leq d(x_1, x_2) + d(x_2, x_3)$

To so **minimalni pogoji za smiselnost**. Če izbrana razdalja ne zadošča tem pogojem, se algoritmi ne obnašajo smiselno. 

Pomembna lastnost za nas je
$$ x_1\to x_2  \Rightarrow  d(x_1, x_2) \to 0. $$

Velja tudi, da ko sta si $x_1$ in $x_2$ bolj podobna, je $d(x_1, x_2)$ manjša.

## ■ Primeri razdalj

Vzemimo, da ja $\vec{x}=x(h)$ vektor lastnosti entitete $h\in H$ oziroma $\vec{x}=x(u)$ vektor lastnosti uporabnika $u\in U$. Kako te lastnosti določamo, obravnavamo kasneje. 


<br>

Pomembni **primeri** razdalj so

**(1) Kosinusna razdalja:** za $\vec{x}_1\ne 0$ in $\vec{x}_2\ne 0$
$$ d(\vec{x}_1, \vec{x}_2) = \frac{\vec{x}_1.\vec{x}_2}{\|\vec{x}_1\|.\|\vec{x}_2\|}. $$
Ta razdalja meri le kosinus kota oz. smer med vektorjema. 

<br>

**(2) p-norma:** za $p\geq 1$
$$ d(\vec{x}_1, \vec{x}_2) = \|\vec{x}_1 - \vec{x}_2\|_p = \sqrt[p]{\sum_i |x_{1i} - x_{2i}|^p}. $$
Parameter $p$ določa osnovne lastnosti 
- $p=1$: vsota razlik, izmeri veliko število malih razlik
- $p=2$: Evklidska razdalja, ravnotežje med obema
- $p=\infty$: maksimalna razlika, izmeri največjo razliko v lastnostih


Bistveno je tudi, **kako** so značilke $x$ določene.

## ■ Matrična faktorizacija in latentne spremenljivke (1)

Zaradi vloge v priporčilnih sistemih bomo pobliže spoznali le matrično faktorizacijo, ki je sedaj glavni pristop v priporočilnih sistemih. 

_Literatura:_ [Y. Koren, R. Bell, C. Volinsky: MATRIX FACTORIZATION TECHNIQUES FOR RECOMMENDER SYSTEMS, IEEE Computer Society, 2009.](https://ieeexplore.ieee.org/document/5197422)


- uporabniki $u\in U$, vsebine $h\in H$
- podatkovna matrika ocen uporabnikov - večinoma ima **neznane vrednosti** (n.d. = no data)
$$ D=[r_{uh}]_{u\in U, h\in H} $$



- uporabimo oznako $P_{(u)}$ za $u$-to vrstico matrike $P$ in $Q_{(h)}$ za $h$-to vrstico matrike $Q$. 



- Kaj je faktorizacija matrik: $Q.R$, ...



- pojem latentnih značilk (lastnosti)
    - značilke
    - latentne značilke



- Matrična faktorizacija 
    - izberemo število značilk $M$, ki je lahko poljubno, tipično $M\in \{3,\ldots,10\}$. 
    - faktorizacija je razcep na dve matriki oblike
$$ P.Q^\top = D $$

- večino vrednosti je manjkajočih

$$ 
\begin{bmatrix}
 p_{11} & \ldots & p_{1M} \cr
 \vdots & \ddots & \vdots \cr
 \vdots & \ddots & \vdots \cr
 \vdots & \ddots & \vdots \cr
 p_{n1} & \ldots & p_{nM}
\end{bmatrix}
.\begin{bmatrix}
 q_{11} & \ldots & \ldots & \ldots & q_{M1} \cr
 \vdots & \ddots & \ldots & \ldots & \vdots \cr
 q_{M1} & \ldots & \ldots & \ldots &p_{Mm}
\end{bmatrix}
 = 
\begin{bmatrix}
 n.d. & \ldots & \ldots & \ldots & n.d. \cr
 \vdots & \vdots & \vdots & \vdots & \vdots  \cr
 \vdots & \vdots & r_{uh} & \vdots & \vdots  \cr
 \vdots & \vdots & \vdots & \vdots & \vdots  \cr
 n.d. & \ldots & \ldots & \ldots & n.d.
\end{bmatrix} 
$$

## ■ Matrična faktorizacija in latentne spremenljivke (2)



#### Osnovno sklepanje

- Matriki $P$ in $Q$ določimo tako, da enakost velja pri **znanih vrednostih ocen** matrike $D$. S tem so določene vse vrednosti matrik $P$ in $Q$. 
- Predpostavka: če $P$ in $Q$ pravilno predstavi znane ocene uporabnikov, **bi pravilno ocenila tudi neznane ocen.** S tem ugotovimo **kako bi uprabnik ocenil vsebine, ki jih še ni videl.**
- Ker vrstice $P$ pripadajo posameznim upoabnikom in "delujejo" za vse vsebine $h\in H$, je $P_{(u)}$ vektor latentnih lastnosti uporabnika $u$
- Ker vrstice $Q$ pripadajo posameznim vsebinam in "delujejo" za vse uporabnike $u\in U$, je $Q_{(h)}$ vektor latentnih lastnosti vsebine $h$ 


#### Kako deluje
1. S pomočjo ocenjenih matrik $P$ in $Q$ določimo manjkajoče vrednosti matrike $D$, to je manjkajoče ocene uporabnikov
$$ P.Q^\top = \hat{D} $$
2. Te ocene uporabimo tako, da jih uredimo po padajočem vrsnem redu in jih uporabniku predstavimo npr. prvih $15$;

## ■ Podobnost in razdalja


Podobnost 
$$ s: X\times X \to ℝ_+, $$
ni tako natančno definirana kot razdalja. Osnovna lastnost podobnosti je, da če sta $\vec{x}_1$ in $\vec{x}_2$ bolj podobna, je $s(\vec{x}_1, \vec{x}_2)$ večja. 


Podobnost meri isto lastnost kot razdalja, le v **obrnjeni smeri**. Primeri povezava med simetričnostjo in razdaljo so

**(1) Obrnjena razdalja**
$$  s(\vec{x}_1, \vec{x}_2) =  \frac{1}{d(\vec{x}_1, \vec{x}_2)} $$
Ta simetrija doseže $\infty$ ko razdalja pade na $0$


**(2) Obrnjena razdalja s popravkom**
$$  s(\vec{x}_1, \vec{x}_2) = \frac{1}{1+d(\vec{x}_1, \vec{x}_2))} $$
Ta simetrija slika na interval $[0,1]$ in ko pade razdalja med vektorjema na $0$, doseže vrednost $1$.


**(3) Eksponentna**
$$  s(\vec{x}_1, \vec{x}_2) = e^{-\lambda d(\vec{x}_1, \vec{x}_2)} $$
Ta simetrija slika na interval $[0,1]$ in ko pade razdalja med vektorjema na $0$, doseže vrednost $1$. Parameter $\lambda$ regulira hitrost naraščanja simetrije proti $1$ ko se vektorja bližata v isti vektor, torej njuna razdalja pada na $0$.
