# ■ Introduction: what is an instrument and what is a construct

#### Instrument

An instrument is a list of questions or tasks that the user answers before, during, or after a session of interaction with the system.

An example of the instrument and construct is in the picture.

<img style="float: center; width: 200px; margin: 20px 20px 50px 0;" src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/Instrument-NasaTlx.png">

Questions can be submitted electronically or on paper. Tasks are given graphically or otherwise.
An example of an instrument that is not in the form of questions is the FeelTrace instrument for capturing affective and consequently emotional states.

<img style="float: center; width: 250px; margin: 20px 20px 50px 0;" src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/FeelTraceInstr.png">





#### Construct


A construct is a label for a given domain of behavior or property. We are interested in behaviors or properties relevant to user interaction with smart systems. Constructs usually consist of several dimensions, they indicate individual aspects of properties or behavior and are given by individual numbers.

For example the Big personality construct (see below) describes the basic characteristics of the user's personality and consists of five dimensions. Each covers its own aspect and its intensity is given by a number between 0 and 100.

# ■ Load Measurement: NASA TLX, construct

#### Construct: Six-Dimension Load Model


|Dimenzija|opis|
|:--- |:--- |

```python
import numpy as np
import pandas as pd 
import sklearn as sk
from sklearn.preprocessing import MinMaxScaler



# @brief get construct scores from users' anwsers
# @par qa_mat matrix of user anwsers
# @par la_mat loading matrics 
# @par 
# @return score matrix
def get_scores(qa_mat, la_mat, meth_code='inv'):

    # Compute it     
    if meth_code=='inv':
        raw_S = qa_mat @ np.linalg.pinv(la_mat).T
    
    if meth_code=='cov':
        raw_S = qa_mat @ np.cov(qa_mat.T) @ la_mat

    # Scale it 1 - 10
    scaler = MinMaxScaler(feature_range=(1, 10))
    scaler.fit(raw_S)
    S = scaler.transform(raw_S)

    return S

```

# ■ Calculation of construct or scale points

### Statistical evaluation of a construct or scale


##### User responses

Matrix of responses of users $U$ to questions $Q$,
$$ X = [a_{uq}], \qquad u\in U, q\in Q, $$
where $U$ is the set of test users and Q is the set of their answers to the questions.

##### The model is given by a matrix of weights

Weight matrix
$$ L = [\lambda_{qi}], \qquad q\in Q, i\in F, $$
where Q is the set of their answers to the questions and $F$ is the set of factors.


##### Calculation of points (English sores)

The point matrix $S=[s_{ui}], u\in U, i\in F$ given by
$$ S = X.C^{\top}.L, $$
where $C$ is the covariance matrix of the responses $X$.




### Simplifications: average of responses

Dimension values are averages of responses. This is a simplification of the earlier formula where we assume that all covariances are equal and all factor weights are equal.

# ■ The problem of questionnaires and measurement characteristics

#### Problem

Questionnaires can give **useless results** because the answers to the questions are not related to the real situation. There are several reasons for this:
1. The concepts in the questions or the questions themselves are **not well defined**: expressions in the human language can carry several meanings at the same time and if these are not specified, the answers refer to different concepts, objects, etc.
2. It is not clear in the question what it refers to - the reference is not given: the question may refer to the quality of the service, interface, content, etc.;
3. The questions may be offensive or disturbing to the test subject
As a result, it is necessary to be skeptical about the applicability of individual questions.

#### Solution

Measurement characteristics detect questions that **do not perform well enough**, i.e. do not give meaningful answers. The point is that the measurement characteristics detect the properties of the answers that must always be valid (e.g. stability with multiple questions, etc.) and thus good measurement characteristics are a **necessary condition for the applicability of the answers**, but not a sufficient condition.

# ■ Reliability

#### Def.: Reliability is the overall consistency of a measure. A measure is said to have a high reliability if it produces similar results under consistent conditions.


#### Types of Reliability
1. Reliability over time - test-retest reliability: whether repeated testing gives similar results (_not applicable for the example of two-way conversation with Alexa_)
2. Internal consistency: are the answers between the questions sufficiently connected (_essential for the example of a two-way conversation with Alexa_ - see below)
3. Consistency between researches (researchers) (_not applicable for the case of two-way conversation with Alexa_)


#### Internal consistency - method of parallel forms (split half)

Answers the question **Are the answers/data stable?**

##### Determination process
$\def\s{{\sigma}}$
In this method for given data:
1. The test subjects are randomly divided into two halves
2. We calculate the correlations between them $r_{12}$
3. We calculate the reliability coefficient according to the Spearman-Brown formula
    - for default equal standard deviations
   $$ r_{tt} = \frac{2 r_{12}}{1 + r_{12}} $$
    - for different standard deviations
   $$ r_{tt} = \frac{4 \s_1 \s_2 r_{12}}{\s_1^2 + \s_2^2 + 2 \s_1 \s_2 r_{12}}, $$
   where $\s_1$ and $\s_2$ are the standard deviations of individual data.



#### Internal consistency - Cronbach alpha

Answers the question **Do all the questions/data together measure a coherent concept?**

##### Formula
The results (answers) to $K$ questions $X_i$ are added up in
$$ Y = X_1 + \cdots + X_K, $$
calculate the standard deviations of the individual results $\s_{X_i}$ and the sum $\sigma_{Y}$, and calculate
$$ \alpha = \frac{K-1}{K} \left(1-\frac{\sum _{i=1}^K\s_{X_i}^2}{\s_Y^2}\right) $$

```python
## Measurement characteristics
import numpy as np
import pandas as pd 
import random

# Load data 
post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/2020-2021_PredPost-vprasalnik(dvogovor)(Responses).csv'
#post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/PostQs_Responses.csv'
data_df = pd.read_csv(post_fn, header=0, sep=';', encoding='utf8').drop([13])


# Selectors 
pre_qs_inds = list(range(5,11))
post_qs_inds = list(range(11,36))


# ==============================================================================================
## Validitiy
# We estimate is OK

# ==============================================================================================
## Reliability

# Parallel forms - inter method reliability: divide test tasks to two parts and corelate anwsers 
# Split half
np.random.shuffle(post_qs_inds)
perm_inds_1, perm_inds_2 = post_qs_inds[0:12], post_qs_inds[12:24] 
anws_1, anws_2 = data_df.iloc[:, perm_inds_1], data_df.iloc[:, perm_inds_2]


# Get correlation coefficient
half1_pd = anws_1.sum(axis=1)
half2_pd = anws_2.sum(axis=1)
corr = np.corrcoef(half1_pd, half2_pd)[0,1]


# ==============================================================================================
print ('====================================================================================')
print ('== For instruments')
# Spearman–Brown prediction formula
# Equal variances
req_tt = 2*corr / (1.0 + corr)
print ('Equal var, r_tt=', req_tt)

# Non-equal variances
sd1, sd2 = np.std(half1_pd), np.std(half2_pd) 
rneq_tt = 4*sd1*sd2*corr / (sd1**2 + sd2**2 +  2*sd1*sd2*corr)
print ('Std 1 = ', sd1)
print ('Std 2 = ', sd2)
print ('Not equal var, r_tt=', rneq_tt)


## ==============================================================
# Internal consistency - Cronbach alpha
# Verifies the instrument as a whole

from reliabilipy import reliability_analysis

# Cronbach alpha
#ra = reliability_analysis(raw_dataset=qa_mat, is_corr_matrix=False, impute='median')
#ra.fit()
#print ('Cronbach alpha: ', ra.alpha_cronbach)


post_qs_df = data_df.iloc[:, 11:36]
_,K = post_qs_df.shape
X = post_qs_df.sum(axis=1)
ss_X = np.var(X, ddof=1)
ss_Yi = np.var(post_qs_df, ddof=1)
ss_Y = ss_Yi.sum()

Cronb_a = (K/(K-1))*(1.0 - ss_Y/ss_X)
print ('Cronbach Alpha =', Cronb_a)
```
