## ■ Ovrednotenje: P, R, A, F, AUC

Namen vrednotenje je približati oceno uspešnosti v realni aplikaciji. 

$\def\TP{{\text{TP}}}$
$\def\TN{{\text{TN}}}$
$\def\FP{{\text{FP}}}$
$\def\FN{{\text{FN}}}$
$\def\TPR{{\text{TPR}}}$
$\def\FPR{{\text{FPR}}}$

### Matrika razvrščanja:
Definiramo za binarno klasifikacijo (dva razreda). Za več razredov ocenjujemo posamezen razred proti ostalim. 
$$ M = 
\begin{array}{c|c|c|}
 & \hat{C} & \hat{\neg C} \\ \hline
 C & \TP  & \FN \\ \hline
 \neg C  & \FP & \TN \\ \hline
\end{array}
$$

Pri tem je 
- TP: pravilno pozitiven 
- TN: pravilno negativen
- FP: napačno pozitiven
- TN: pravilno negativen 


### Mere uspešnosti
- natančnost (accuracy) $A = \frac{\TP+\TN}{\TP+\TN+\FP+\FN}$
- priklic (recall) $R = \TPR = \frac{\TP}{\TP+\FN}$, true positive rate
- točnost (precission) $P = \frac{\TP}{\TP+\FP}$, 
- lažno pozitivni $\FPR = \frac{\FP}{\FP+\TN}$, false positive rate
- F-mera $F = 2\frac{P\cdot R}{P + R}$

Opažanja:
- meri P in R sami zase nista uporabni
- neuspešen razvrščevalnik ima lahko visoko uspešnost! Zato tekmujemo z naključjem, ki NI $P=R=0.5$! Izračunamo matriko razvrščanja naključnega klasifikatorja in se primerjamo z njo!


#### ROC krivulja in povoršina pod njo (AUC)

Izvor krivulje je dekodiranje signala s šumom, kar je ekvivalentno s klasifikatorjem s točkami (angl. scoring classifier). 

ROC (angl. receiver operating curve) je krivulja $\TPR=f(\FPR)$. Ploščino pod ROC krivuljo označimo z AUC. Izkaže se, da je enaka verjetrnosti, da klasifikator pravilno razvrsti naključno izbran element.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_predict, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score

# 1. Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Binarize the output for the ROC curve (convert multi-class to binary classification)
y_binary = label_binarize(y, classes=[0, 1, 2])

# 2. Initialize the Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# 3. Perform 10-fold Cross-Validation
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Get cross-validated predictions (labels)
cv_predictions = cross_val_predict(clf, X, y, cv=cv)

# Get cross-validated predicted probabilities (needed for ROC curve)
cv_probabilities = cross_val_predict(clf, X, y, cv=cv, method='predict_proba')

# 4. Compute Precision, Recall, and F1-score
precision = precision_score(y, cv_predictions, average='weighted')
recall = recall_score(y, cv_predictions, average='weighted')
f1 = f1_score(y, cv_predictions, average='weighted')

print(f'Precision (10-fold CV): {precision:.2f}')
print(f'Recall (10-fold CV): {recall:.2f}')
print(f'F1-score (10-fold CV): {f1:.2f}')

# 5. Compute ROC Curve using cross-validation probabilities (one-vs-rest approach)
fpr, tpr, _ = roc_curve(y_binary[:, 1], cv_probabilities[:, 1])  # Choosing class 1 for demonstration
roc_auc = auc(fpr, tpr)

# 6. Plot the ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line for reference
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Random Forest Classifier (10-Fold CV)')
plt.legend(loc='lower right')
plt.show()


```

## ■ Konfiguracija: dendrogram in komolčna krivulja


### Dendrogram

Dendrogram je drevo, ki prikazuje združevanje bližnjih razredov (na začetku elementov) v nove večje razrede, vse do enega samega razreda. 

Proti koncu nam ta struktura pomaga določiti smiselno število razredov.


### Komolčna krivulja

Komolčna krivulja podaja vsoto notranjih razdalj med razredi v odvisnoti od števila razredov. Torej sledi združevanje razredov dendrograma. 

<p style="margin-bottom:1cm;"></p><br><br><br>
<div style="width:100%;text-align:right;font-weight:bold;font-size:1.2em;"> 14 </div>
<img src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/footer_full.jpg">

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate sample data
np.random.seed(42)
X, _ = make_blobs(n_samples=300, centers=4, n_features=2, cluster_std=1.0, random_state=42)
X_df = pd.DataFrame(X, columns=['f1', 'f2'])

# Elbow Method to find optimal K
wcss = []  # Within-cluster sum of squares
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(10, 4))
plt.plot(k_range, wcss, marker='o', linestyle='--', color='b')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Within-cluster sum of squares)')
plt.title('Elbow Method for Optimal K')
plt.show()

# Dendrogram for Hierarchical Clustering
plt.figure(figsize=(10, 5))
linkage_matrix = linkage(X, method='ward')
dendrogram(linkage_matrix)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()

# Apply K-Means with optimal K (from the elbow method, assume K=4)
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
X_df['Cluster'] = kmeans.fit_predict(X)

# Compute cluster means
cluster_means_df = X_df.groupby("Cluster").mean()
#print("Cluster Means:")
#print(cluster_means_df)
cmap = plt.cm.RdYlGn
fig, axs = plt.subplots(optimal_k, figsize=(4,2*optimal_k))
fig.suptitle('Cluster means')
for ii in range(optimal_k):
    axs[ii].bar(X_df.columns[:2], cluster_means_df.iloc[ii], color=['r', 'g'])
    axs[ii].set_title('Cluster '+str(ii+1))
fig.tight_layout()

# Plot K-Means clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=X_df['Cluster'], palette="viridis", s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            c='red', marker='X', s=200, label="Centroids")
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```

## ■ Regresijski model

$\def\b{{\beta}}$

Linearna regresija
$$ y = \b_0 + \b_1 x_1 + \b_2 x_2 + \cdots  $$

Algoritem za določanje kodeficientov $\b_i$: minimizira kvadratično napako - optimizacijski problem, ki je rešliv z matričnim računom.

Privzetki:
- linearne odnos med neodvisnimi in odvisnimi spremenljivkami
- normalna porazdeljenost ostankov $y_i - \hat{y_i}$
- homogenost - enaka variabilnost pri različnih vrednostih neodvinih spremenljivk

## ■ Ovrednotenje regresijskega modela: $R^2$

$\def\b{{\beta}}$
$\def\ovr#1{{\overline{#1}}}$

Kvaliteto prileganja regresijskega modela 
$$ y = \b_0 + \b_1 x_1 + \b_2 x_2 + \cdots  $$
ocenimo z determinacijskim koeficientom $R^2$, ki je delež pojasnjene variance
$$ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $$

Pri tem je 
- celotna variabilnost: $SS_{tot} = \sum_i (y_i - \ovr{y})^2$
- preostala variabilnost: $SS_{res} = \sum_i (y_i - \hat{y})^2$





<p style="margin-bottom:1cm;"></p><br><br><br>
<div style="width:100%;text-align:right;font-weight:bold;font-size:1.2em;"> 14 </div>

```python
#
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generate random data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # Feature variable
y = 4 + 3 * X + np.random.randn(100, 1)  # Target with some noise

# Train the model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Calculate R² score
r2 = r2_score(y, y_pred)
print(f"R² Score: {r2:.4f}")

# Plot the data and regression line
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression Line")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.title("Linear Regression with R² Score")
plt.legend()
plt.show()
```
