## ■ Umestitev

Modeliranje uporabnikv omogoča vnos in upoštevanje specifične informacije posameznega uporabnika z namenom optimizacije storitve, ki navadno nastopi v obliki prilagajanja storitve. Jasno je, da storitev brez uporabnih informacij o posameznem uporabniku ne more te storitve optimalno prilagoditi temu uporabniku. 

Najširša informacija o uporabniku je pripadnost segmentu uporabnikov. Uporabnike storitve uvrstimo v specifične razrede, ki jih imenujemo segmente - rezultat postopka segmentacije. Ti postopki posameznega uporabnika uvrstijo v okvir ostalih uporabnikov. Postopke segmentacije so razvili na povezanih področjih trženja, a so uporabni tudi za okvir optimizacije z uporabniki. 

Postopke prilagajanja sami osebi neodvisno od ostalih uporabnikov imenujemo personalizacija. Vključuje informacije o uporabniki kot je osebnost in zgodovina njegovih aktivnosti, zbrana v uporabnikovem modelu. 

Študije kažejo, da je situacija uporabnika, v kateri se nahaja med samo uporabo storitve, ključna za optimizacijo in navadno pomembnejša od personalizacije. Situacijo uporabnika imenujemo kontekst. Opišemo ga z drevesno struktiro, pri čemer prva razvejitev zajema tri komponente, ki so čas (npr. ura v dnevu, dan v tednu, ipd), kraj (npr. doma, na poti, v službi) ter aktivnost (kaj uporabnik počne med samo uporabo aplikacije). Z razpoznavo in upoštevanjem konteksta dosežemo situacijsko zavedanje storitve. 

Modeliranje uporabnikov pokriva za optimizacijo koristne informacije o uporabnikovi pretekli uporabi in odzivih na uporabo v okviru celoten skupine uporabnikov. Algoritmi modeliranja uporabnikov so zelo raznoliki, tu pokrivamo Markovske verige (s lsučajnostjo povezano zaporedje dogodkov/aktivnosti) ter latnetno modeliranje s pristopi matrične faktorizacije.
